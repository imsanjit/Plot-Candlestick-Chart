```python
import pandas as pd
import mplfinance as mpf

%matplotlib inline
```

```python
df = pd.read_csv('AMZN.csv')
df.head()
```

```python
df.info()
```

```python
df.columns
```

```python
df['Date'] = pd.to_datetime(df['Date'] )
df.info()
```

```python
df = df.set_index('Date')
```

```python
df.head()
```

```python
mpf.plot(df)
```

```python
mpf.plot(df, type='line', volume=True)
```

```python
mpf.plot(df['2020-03'], volume=True)
```

```python
mpf.plot(df['2020-03'], type='candle', volume=True)
```

```python
mpf.plot(df['2020-03': '2020-07'], type='candle', mav=(20), volume=True)
```

```python
mpf.plot(df['2020-03': '2020-07'], 
         type='candle', 
         mav=(20), 
         volume=True,
         tight_layout=True)
```

```python
mpf.plot(df['2020-03': '2020-07'], type='candle', 
         mav=(20), volume=True, tight_layout=True,
         figratio=(20,10),
         title='Amazon price 2020')
```

```python
mpf.plot(df['2020-03': '2020-07'], type='candle', 
         mav=(20), volume=True, tight_layout=True,
         figratio=(20,10),
         title='Amazon price 2020',
         style='yahoo')
```
